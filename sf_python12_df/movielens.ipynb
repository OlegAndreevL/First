```python
import pandas as pd
dates = pd.read_csv('movielens/dates.csv', sep=',')
movies = pd.read_csv('movielens/movies.csv', sep=',')
ratings1 = pd.read_csv('movielens/ratings1.csv', sep=',')
ratings2 = pd.read_csv('movielens/ratings2.csv', sep=',')
dates['date'] = pd.to_datetime(dates['date'])
years = dates['date'].dt.year
ratings = pd.concat([ratings1, ratings2], ignore_index=True)
ratings = ratings.drop_duplicates(ignore_index=True)
ratings_dates = pd.concat([ratings, dates], axis=1)
joined = ratings_dates.join(movies.set_index('movieId'), on='movieId', how='inner')
merged = ratings_dates.merge(movies, on='movieId', how='left')
display(merged)
```
